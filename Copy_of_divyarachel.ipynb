```python
#2nd part of assignment Get_to_know_your_data
import pandas as pd
import numpy as np
iris_df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
iris_df.head(10)#viewing first 10 rows
print(iris_df.shape)

print(iris_df.shape[0])
print(iris_df.shape[1])
print(iris_df.columns)#printing columns
#checking index
print(iris_df.index)


print(iris_df.head())

new_index = np.random.permutation(iris_df.index)
iris_df = iris_df.reindex(index = new_index)

print(iris_df.head())

#original

print(iris_df.head())

iris_df['sepal_width'] *= 10

#changed

print(iris_df.head())

#lets undo the operation

iris_df['sepal_width'] /= 10

print(iris_df.head())
iris_df[iris_df['sepal_width']>3.3]
iris_df[(iris_df['sepal_width']>3.3) & (iris_df['species'] == 'versicolor')] 

iris_df.sort_values(by='sepal_width')#for ascending
iris_df.sort_values(by='sepal_width',ascending=False)#for descending

species = iris_df['species'].unique()
print(species)

setosa = iris_df[iris_df['species'] == species[0]]
setosa.head()

#for 2 other species
versicolor = iris_df[iris_df['species'] == species[1]]
versicolor.head()
virginica = iris_df[iris_df['species'] == species[2]]
virginica.head()

#description of each species
print(setosa.describe())
print(versicolor.describe())
print(virginica.describe())
```
