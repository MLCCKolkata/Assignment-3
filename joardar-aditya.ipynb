```python
wine_df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data')
wine_df
```

```python
wine_df.head(5)
```

```python
wine_df_copy = wine_df
odd_values = []
for i in range(wine_df_copy.shape[0]) :
  if i%2 == 1 :
    odd_values.append(i)
wine_df_copy.drop(odd_values,0)
```

```python
wine_df_copy.columns = ['1','Alcohol',' Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols','Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline']
wine_df_copy
```

```python
wine_df_copy.loc[0:2,'Alcohol'] = np.nan
wine_df_copy
```
