#### Create Series from numpy array, list and dict

Don't know what a series is?

[Series Doc](https://pandas.pydata.org/pandas-docs/version/0.22/generated/pandas.Series.html)

```python
a_ascii = ord('A')
z_ascii = ord('Z')
alphabets = [chr(i) for i in range(a_ascii, z_ascii+1)]

print(alphabets)

numbers = np.arange(26)

print(numbers)

print(type(alphabets), type(numbers))

alpha_numbers = dict(zip(alphabets, numbers))

print(alpha_numbers)

print(type(alpha_numbers))
```

```python
series1 = pd.Series(alphabets)
print(series1)
```

```python
series2 = pd.Series(numbers)
print(series2)
```

```python
series3 = pd.Series(alpha_numbers)
print(series3)
```

#### Extract Items from a series

```python
ser = pd.Series(list('abcdefghijklmnopqrstuvwxyz'))
pos = [0, 4, 8, 14, 20]

vowels = ser.take(pos)

df = pd.DataFrame(vowels)#, columns=['vowels'])

df.columns = ['vowels']

df.index = [0, 1, 2, 3, 4]

df
```

```python
#I have used the mathematics to draw the regression line of alcohol vs colour intensity
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data', sep=',')
perm=np.random.permutation(data.index)
data=data.reindex(perm)
data.columns="Class Alcohol Malic_acid Ash Alcalinity_of_ash Magnesium Total_phenols Flavanoids Nonflavanoid_phenols Proanthocyanins Color_intensity Hue OD280/OD315_of_diluted_wines Proline".split()
```

```python
data.describe()
```

```python
Y=data['Alcohol']
X=data['Color_intensity']
#calculating the mean of all values of X and Y
mean_x=np.mean(X)
mean_y=np.mean(Y)
#total number of values
n=len(X)
num=0
den=0
for i in range(n):
    num+=(X[i]-mean_x)*(Y[i]-mean_y)
    den+=(X[i]-mean_x)**2
m=num/den
c=mean_y-mean_x*m
#print the values of m and c
print("y= {0} x + {1}".format(m,c))
```

```python
#plotting values and the regression line
max_x=np.max(X)+100
min_x=np.min(X)-100

#calculating the line values of x and y
x=np.linspace(min_x,max_x,1000)
y=m*x+c

#plotting the line
plt.plot(x[450:550],y[450:550],color='#58b970',label='Regression Line')

#plotting the scatter points
plt.plot(X,Y,'bo',label='Scatter Plot')

plt.xlabel('total_rooms')
plt.ylabel('median_house_value')
plt.title('Median House Value vs Total Rooms')
plt.show()
```
