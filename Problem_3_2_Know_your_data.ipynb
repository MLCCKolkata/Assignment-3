Understanding Data

Importing necessary modules

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading CSV data to  DataFrame

```
iris_df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

Viweing the top 5 *rows*

```
iris_df.head()
```

Finding the number of rows and columns

```
print(iris_df.shape)


print(iris_df.shape[0])
print(iris_df.shape[1])
```

Printing all columns

```
print(iris_df.columns)
```

Checking Index

```
print(iris_df.index)
```

Shuffling Data

```
print(iris_df.head())

new_index = np.random.permutation(iris_df.index)
iris_df = iris_df.reindex(index = new_index)

print(iris_df.head())
```

Applying an operation on the whole column

```
#original data

print(iris_df.head())

#changing the data

iris_df['sepal_width'] *= 10

#changed data

print(iris_df.head())

#Undoing the operation

iris_df['sepal_width'] /= 10

print(iris_df.head())
```

Showing all the rows where sepal_width > 3.3

```
iris_df[iris_df['sepal_width']>3.3]
```

Clubbing two filters together - Find all samples where sepal_width > 3.3 and species : "versicolor"


```
iris_df[(iris_df['sepal_width']>3.3) & (iris_df['species'] == 'versicolor')] 
```

Sorting a column by value

```
iris_df.sort_values(by='sepal_width')# ascending order
```

```
iris_df.sort_values(by='sepal_width', ascending = False)
```

Listing all unique series

```
species = iris_df['species'].unique()

print(species)
```

Selection using boolean mask

```
setosa = iris_df[iris_df['species'] == species[0]]

setosa.head()
```

```
versicolor = iris_df[iris_df['species'] == species[1]]

versicolor.head()
```

```
virginica = iris_df[iris_df['species'] == species[2]]

virginica.head()
```

Describing each species

```
setosa.describe()
```

```
versicolor.describe()
```

```
virginica.describe()
```

Plotting the data

```
plt.hist(setosa['sepal_length'])
plt.hist(versicolor['sepal_length'])
plt.hist(virginica['sepal_length'])
```
